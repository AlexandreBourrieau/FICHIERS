### Interface de modification

Cette interface va permettre de caractériser les labels suivants sur chaque photos du dataset :  
- Tête : Correct (C) ou Non-Correct (NC)  
- Bras droit : Correct (C) ou Non-Correct (NC)  
- Bras gauche : Correct (C) ou Non-Correct (NC)  
- Jambe droite : Correct (C) ou Non-Correct (NC)  
- Jambe gauche : Correct (C) ou Non-Correct (NC)

Le format du nom de fichier est le suivant : id_TETE_BD_BG_JD_JG.jpg  

Par exemple :  
- ID_C_C_C_C_C.jpg : Tout est correct (C)  
- ID_C_NC_NC_C_C.jpg : Tête C / bras droit NC / bras gauche NC / jambe droite C / jambe gauche C
  

On commence par créer tous les fichiers "CORRECTS" :

```python
import os
from os import listdir
from os.path import isfile, join
import shutil

nom_position = "AssisesBrasTendus"
CATEGORIES = ['correct', 'non_correct']

repertoire_courant = os.getcwd()
repertoire_C = repertoire_courant + "/dataset/" + nom_position + "/CORRECT/"

# Récupère toutes les images du répertoire CORRECT
fichiers_images_corrects = [f for f in listdir(repertoire_C) if f.split(".")[1]=="jpg" and len(f.split("_"))==1]

# Enregistre les nouvelles images au format id_C_C_C_C_C.jpg
for nom_fichier in fichiers_images_corrects:
    shutil.copyfile(repertoire_C+nom_fichier,repertoire_C+nom_fichier.split(".")[0]+"_C_C_C_C_C.jpg")
```
