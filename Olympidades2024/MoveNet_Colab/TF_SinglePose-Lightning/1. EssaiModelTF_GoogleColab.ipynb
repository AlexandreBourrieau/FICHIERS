```python
import tensorflow as tf

model_path = "SavedModel"
model = tf.saved_model.load(model_path)
```

```python
from PIL import Image
import numpy as np
from io import BytesIO

img_data = tf.io.gfile.GFile('/content/58959510-concept-de-yoga-et-de-remise-en-forme-de-grossesse-portrait-de-beau-jeune-yoga-enceinte-mod√®le.jpg', 'rb').read()
image = Image.open(BytesIO(img_data))
image = image.resize((192,192))
image_np  = np.array(image.getdata()).reshape((192, 192, 3)).astype(np.int32)
input_tensor = tf.convert_to_tensor(np.expand_dims(image_np, 0))
```
